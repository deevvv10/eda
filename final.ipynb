```python
# Load required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.decomposition import PCA

# Read CSV file from GitHub
url = "https://raw.githubusercontent.com/salemprakash/EDA/main/Data/Fertility2.csv"
df = pd.read_csv(url)

# View the first few rows
print(df.head())

# Data Cleaning
print(df.info())
df.columns = df.columns.str.replace(" ", "_").str.strip()
df = df.drop(columns=['rownames'], errors='ignore')
print(df.isnull().sum())
df = df.drop_duplicates()

# Convert data types
factor_columns = ['morekids', 'gender1', 'gender2', 'afam', 'hispanic', 'other']
for col in factor_columns:
    df[col] = df[col].astype('category')

df['work'] = pd.to_numeric(df['work'], errors='coerce')
df['age'] = pd.to_numeric(df['age'], errors='coerce')
df = df.dropna()

# Summary after cleaning
print(df.describe(include='all'))
print(df.info())

# Create Derived Variables
def age_group(age):
    if age < 25:
        return "Under 25"
    elif 25 <= age <= 35:
        return "25 to 35"
    else:
        return "Above 35"


df['age_group'] = df['age'].apply(age_group)
df['minority'] = np.where((df['afam'] == 'yes') | (df['hispanic'] == 'yes') | (df['other'] == 'yes'), 'yes', 'no')

heavy_workers = df[df['work'] > 40]
more_kids = df[df['morekids'] == 'yes']

# Summarizing Data
print(df.groupby('gender1')['work'].mean())
print(df.groupby(['age_group', 'morekids']).size().reset_index(name='count'))
print(df.groupby('minority')['work'].mean())

print(df.sort_values(by='age', ascending=False).head())
print(df.sort_values(by=['age', 'work'], ascending=[False, False]).head())

# Save cleaned data
df.to_csv("Cleaned_Fertility2.csv", index=False)

# 1D Visualization
sns.histplot(df['age'], bins=20, kde=False, color='skyblue')
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

sns.countplot(x='morekids', data=df, color='tomato')
plt.title("Count of Families with More Kids")
plt.xlabel("More Kids")
plt.ylabel("Count")
plt.show()

sns.boxplot(y='work', data=df, color='orange')
plt.title("Boxplot of Weekly Work Hours")
plt.ylabel("Work Hours")
plt.show()

# 2D Visualization
sns.scatterplot(x='age', y='work', data=df, color='steelblue')
plt.title("Age vs Work Hours")
plt.xlabel("Age")
plt.ylabel("Work Hours")
plt.show()

sns.boxplot(x='gender1', y='work', data=df)
plt.title("Work Hours by Gender of First Child")
plt.xlabel("Gender1")
plt.ylabel("Work Hours")
plt.show()

sns.countplot(x='age_group', hue='morekids', data=df)
plt.title("Age Group vs Morekids")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.show()

# N-D Visualization
sns.scatterplot(x='age', y='work', hue='morekids', data=df)
plt.title("Age vs Work Hours Colored by Morekids")
plt.show()

sns.relplot(x='age', y='work', col='morekids', row='gender1', data=df, kind='scatter', color='purple')
plt.suptitle("Faceted Plot: Age vs Work by Gender1 and Morekids", y=1.02)
plt.show()

sns.scatterplot(x='age', y='work', size='age', hue='minority', data=df, alpha=0.5)
plt.title("Bubble Plot: Age vs Work Colored by Minority")
plt.show()

# 1D Analysis
print(df['age'].describe())
print(df['morekids'].value_counts())
plt.hist(df['work'], color='skyblue')
plt.title("Histogram of Work Hours")
plt.xlabel("Work Hours")
plt.show()

# 2D Analysis
print(pd.crosstab(df['morekids'], df['gender1']))
sns.boxplot(x='morekids', y='work', data=df, color='tomato')
plt.title("Work Hours by Morekids")
plt.show()

plt.scatter(df['age'], df['work'], color='blue')
plt.title("Age vs Work")
plt.xlabel("Age")
plt.ylabel("Work Hours")
plt.show()

# N-D Analysis
print(df.groupby(['morekids', 'age_group'])['work'].mean().reset_index())
sns.relplot(x='age', y='work', col='morekids', row='gender1', data=df, kind='scatter', color='purple')
plt.suptitle("Age vs Work by Gender1 and Morekids", y=1.02)
plt.show()

# Dimensionality Reduction
label_encoders = {}
for col in df.select_dtypes(include='category').columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

features = df.select_dtypes(include=['int64', 'float64'])
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

pca = PCA()
pca_result = pca.fit_transform(scaled_features)

print(pd.DataFrame(pca.explained_variance_ratio_, columns=["Explained Variance"]))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.title("Scree Plot of Principal Components")
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.show()

plt.figure(figsize=(8, 6))
plt.scatter(pca_result[:, 0], pca_result[:, 1], alpha=0.6, color='blue')
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("2D PCA Projection")
plt.show()

# Model Building
X = df.drop('morekids', axis=1)
y = df['morekids']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

pred_class = model.predict(X_test)

conf_matrix = confusion_matrix(y_test, pred_class)
print(conf_matrix)
print(classification_report(y_test, pred_class))

ConfusionMatrixDisplay(conf_matrix, display_labels=model.classes_).plot(cmap='Blues')
plt.title("Confusion Matrix")
plt.show()
```
